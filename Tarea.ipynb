# Transportar carga con vehículo de CDIO
Para transportar la carga de un vehículo y evaluar si este es capaz de realizarlo se realiza el siguiente procedimiento.

## Inputs
$m_t$: La masa que se quiere trasladar [g]  
$m_v$: Masa del vehículo [g]  
$\mu_f$: Coeficiente de fricción entre el suelo esperado y la rueda del vehículo  
$D$: Distancia a trasladar [m]  


```python
import numpy as np
import matplotlib.pyplot as plt

m_t=1100
m_v=500
mu_f=0.4
D=10
g_0=9.81

M=m_t+m_v
F_f=mu_f*M*g_0

for i in range (1, 10):
    F_thrust=F_f*i
    a=F_thrust/M
    print(f"Thrust: {F_thrust:.2f} N, Acceleration: {a:.2f} m/s^2")
```
