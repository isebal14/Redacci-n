## Escudería 6: Guepardex Racing · CDIO 541352 · Universidad de Concepción

Pipeline de cálculo de primer orden que implementa el diagrama IPO del vehículo radiocontrolado GR001.

---

## Diagrama IPO

```
INPUT                  PROCESS                    OUTPUT
─────────────────────────────────────────────────────────────
m_vehiculo             P1. Masa total              Ft (N)
m_carga        ──▶    P2. Reacciones normales  ──▶ τ_motor (N·m)
µ = 0.515             P3. Roce y tracción          v_max (km/h)
G = 2.588             P4. Torque mecánico          t_carrera (s)
r = 32.5 mm           P5. Velocidad máxima         P_m (W)
V = 11.1 V            P6. Tiempo de carrera        η (%)
RPM = 10 000          P7. Balance eléctrico        consumo (mAh)
                       P8. Consumo y autonomía     Estado sistema
                       s15. Validación
```

---

## Ecuaciones de servilleta

| Proceso | Ecuación |
|---|---|
| Masa total | `M = m_vehiculo + m_carga` |
| Normal trasera | `N_t = M·g·d_CG_del / d_ejes` |
| Roce/tracción | `Fr = µ·N_t  ·  Ft = Fr` |
| Torque rueda | `τ_rueda = (N_t/2)·µ·r` |
| Torque motor | `τ_motor = τ_eje / G` |
| Vel. angular | `ω = 2π·RPM / 60` |
| Vel. máxima | `v_max = (ω_motor / G) · r` |
| Potencia | `P_e = V·I  ·  P_m = τ·ω` |
| Eficiencia | `η = P_m / P_e` |
| Consumo | `E = (I·t / 3600) · 1000 mAh` |
